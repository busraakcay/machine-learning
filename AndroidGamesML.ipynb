```python
y = dataFrame["total_ratings"].values
x = dataFrame.drop("total_ratings",axis=1).values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
#we must split the data set as train and test
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# we must scale the values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
#transforming the values
x_train = scale.fit_transform(x_train)
x_test = scale.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(11,activation="relu"))
model.add(Dense(10,activation="relu"))
model.add(Dense(10,activation="relu"))
model.add(Dense(10,activation="relu"))
model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))

model.add(Dense(1)) #out

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train, y = y_train,validation_data=(x_test,y_test),batch_size=333,epochs=450)
```

```python
lossData = pd.DataFrame(model.history.history)
```

```python
lossData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
predictData = model.predict(x_test)
```

```python
predictData
```

```python
mean_absolute_error(y_test,predictData)
```
